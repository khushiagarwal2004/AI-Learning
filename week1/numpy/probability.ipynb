```python
import numpy as np
```

```python
# simulate probability of getting head
trials = 10000
heads = np.sum(np.random.choice(['H','T'], size=trials) == 'H')
print(heads / trials)
```

```python
# Conditional Probability
A = np.array([1,1,0,1])
B = np.array([1,0,1,1])

P_A_and_B = np.sum((A==1) & (B==1)) / len(A)
P_B = np.sum(B==1) / len(B)

print(P_A_and_B / P_B)
```

```python
# Bernoulli
np.random.binomial(1, 0.5, 10)
```

```python
# Binomial
np.random.binomial(10, 0.5, 100)
```

```python
x = np.array([1,2,3])
p = np.array([0.2,0.5,0.3])

print(np.sum(x*p))
```

```python
data = np.array([1,2,3,4,5])
print(np.var(data), np.std(data))
```

```python
# Gaussian
np.random.normal(0, 1, 1000)
```

```python
# Uniform
np.random.uniform(0, 1, 1000)
```

```python
# Covariance
x = np.array([1,2,3])
y = np.array([2,4,6])
print(np.cov(x,y))
```

```python
# Correlation
print(np.corrcoef(x,y))
```

```python
# Law of Large Numbers
samples = np.random.normal(0,1,10000)
print(np.mean(samples))
```

```python
# Central Limit Theorem
means = [np.mean(np.random.uniform(0,1,100)) for _ in range(1000)]
```

```python
# Maximum Likelihood Estimation (MLE)
data = np.array([1,2,3,4,5])
mu = np.mean(data)  # MLE estimate
print(mu)
```
